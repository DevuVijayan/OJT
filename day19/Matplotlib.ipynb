```python
#PLOT
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#MARKER
x=[1,2,3,4,5]
y=[2,3,6,7,10]
plt.plot(x,y,marker='o',linestyle='-.',color='b',markersize=6,markerfacecolor='r')
plt.title('Line Plot with Markers')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#GRID
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]
plt.plot(x,y1,linestyle='-',color='lime',linewidth=3,marker='o',markersize=8,markerfacecolor='purple')
plt.plot(x,y2,linestyle='--',color='orange',linewidth=3,marker='X',markersize=8,markerfacecolor='red')
plt.title('customized line plot with grid')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('y1,y2')
plt.grid(True)
plt.grid(color='grey',linestyle='--',linewidth=1)
plt.show()
```
