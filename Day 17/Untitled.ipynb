```python
import numpy as np

# 1D array with 8 elements
array_1d = np.arange(1, 9)

# Reshape the array to 3D with dimensions 2x2x2
array_3d = array_1d.reshape((2, 2, 2))

array_3d
```

```python
# Convert the 3D array back to a 1D array
array_1d_again = array_3d.flatten()

array_1d_again
```

```python
import matplotlib.pyplot as plt

# Example data for scatter plot
x = np.random.rand(10)
y = np.random.rand(10)

# Scatter plot
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# Generate 1000 random data points
x_large = np.random.rand(1000)
y_large = np.random.rand(1000)

# Scatter plot with 1000 dots
plt.scatter(x_large, y_large, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with 1000 Dots')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Data points
x_train = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn
y_train = np.array([2, 3, 5, 7, 11])

# Fit the model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict the value of y for x = 6
x_test = np.array([[6]])
y_pred = model.predict(x_test)

y_pred[0]

```
